## Operadores Lógicos

Operador AND (`and`)

<img src="https://aristoteles2pc.files.wordpress.com/2011/03/tabla-de-verdad.png?w=225&h=300" height=225>

```python
# 0 = False, 1 = True
print(False and False)
print(False and True)
print(True and False)
print(True and True)
```

Operador OR (`or`)

<img src="https://aristoteles2pc.files.wordpress.com/2011/03/or1.png?w=228&h=300" height=225>

```python
print(False or False)
print(False or True)
print(False or True)
print(True or True)
```

```python
False or False
```

Operador Negación o NOT (`not`)

<img src="https://aristoteles2pc.files.wordpress.com/2011/03/not1.png" height=180>


```python
print(not False)
print(not True)
```

Operador XOR (`^`)

<img src="https://i.stack.imgur.com/Q7U6I.png" height=150>

```python
print(False ^ False)
print(False ^ True)
print(True ^ False)
print(True ^ True)
```

## Operadores Relacionales

image.png

```python
num_1 = 2
num_2 = 5
num_3 = 2.0001
num_4 = 2
letra_E = 'E'
letra_e = 'e'
```

```python
type(num_1),type(num_2),type(num_3),type(num_4),type(letra_E),type(letra_e)
```

```python
# 2 es mayor que 5?
print(num_1 > num_2)
```

```python
# 2 es menor que 5?
print(num_1 < num_2)
```

```python
# 2 es mayor o igual que 5?
print(num_1 >= num_2)
```

```python
# 2 es menor o igual que 5?
print(num_1 <= num_2)

# 2 es menor o igual que 2?
print(num_1 <= num_4)
```

```python
# 2 es igual a 2.0001?
print(num_1 == num_3)
```

```python
# 2 es diferente a 2.0001?
print(num_1 != num_3)
```

```python
# la letra 'E' es diferente a la letra 'e'?
print(letra_E != letra_e);
```

## Jerarquia de Operadores

| Jerarquia |
|-------------------|
| Paréntesis `( )` |
| Exponente `**`|
| División (`/`) y Módulo (`%`) |
| Multiplicación (`*`) |
| Suma (`+`) y Resta (`-`) |
| Operadores Relacionales (`>`, `<`, `>=`, `<=`, `==`, `!=`) |
| Negación (`not`) |
| Conjunción (`and`) |
| Disyunción inclusiva (`or`) |
| Operadores de Asignación |

*Para más información sobre prioridad de operadores en Python visite este [enlace](https://interactivechaos.com/es/manual/tutorial-de-python/precedencia-de-operadores)*

```python
print( 5 ** (2 * 3) + 3 )
```

```python
print( 5 % 2 * 3 + 3 )
```

```python
print( 5 % (2 * 3) + 3 )
```

```python
a = -5
b = 3
c = 2

print( a / b * a / c > b - a * c )
```

```python
a = 10
b = 7
c = -1

print( a %  b + c * a > 3 or c * c < 5 )
```
